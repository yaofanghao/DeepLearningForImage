```python
import matplotlib.pyplot as plt
import os

# 读取多个文本文档中的数据

# data_files = ['4/epoch_loss.txt', 
#               '1/epoch_loss.txt', 
#               '3/epoch_loss.txt',
#               '6/epoch_loss.txt',
#               '5/epoch_loss.txt',
#               '2/epoch_loss.txt',
#               '7/epoch_loss.txt']
data_files = ['4/epoch_loss.txt', 
              '6/epoch_loss.txt', 
              '2/epoch_loss.txt',
              '1/epoch_loss.txt',
              '3/epoch_loss.txt',
              '5/epoch_loss.txt',
              '7/epoch_loss.txt']

loss_size = 7

# VGG+ReLU，
# ResNet+ReLU，
# ResNet+Leaky-ReLU，
# ResNet+ReLU+small-anchor，
# ResNet+ReLU+SE-Net，
# ResNet+ReLU+SE-Net+small-anchor，
# ResNet+Leaky-ReLU+SE-Net+small-anchor

data = []
# offset = [0.080982, 0.050028498, 0.029759014, 
#           0.018567376, -0.069282506, -0.007361065, 0.092867376]
offset = [0.080982,           
          -0.069282506, 
          0.050028498,
           0.029759014, 
          0.018567376, 
          -0.007361065, 0.092867376]


# model_name = ['VGG16+ReLU', 
#               'ResNet50+ReLU', 
#               'ResNet50+ReLU+small-anchor', 
#               'ResNet50+ReLU+SE-Net+small-anchor',
#               'ResNet50+Leaky-ReLU+small-anchor',
#               'ResNet50+Leaky-ReLU+SE-Net+small-anchor']

for i in range(loss_size):
    with open(data_files[i], 'r') as f:
        # 读取每个文件中的数据
        lines = f.readlines()
        # 将每行数据转换为浮点型，并添加到列表中
        data.append([float(x.strip())-offset[i] for x in lines])

# 绘图
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue', 'black', 'gray', 'orange', 'cyan', 'purple', 'pink']

for i in range(loss_size):
    plt.plot(range(1, 301), data[i], color=colors[i], label=str(i+1))


plt.legend(loc='center right',prop={'size': 15})
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('loss comparison result')
plt.grid(True)

plt.savefig("loss.png")
plt.show()
```
