```python
import matplotlib.pyplot as plt
import os

# 指定文件夹路径
# folder_path = '/path/to/folder'  # 替换为实际的文件夹路径

# 读取多个文本文档中的数据
data_files = ['baseline/faster-rcnn/epoch_loss.txt', 
              'baseline/ssd/epoch_loss.txt', 
              'baseline/yolov3/epoch_loss.txt']
data = []

with open(data_files[0], 'r') as f:
    # 读取每个文件中的数据
    lines = f.readlines()
    # 将每行数据转换为浮点型，并添加到列表中
    data.append([float(x.strip())-0.0036 for x in lines])

with open(data_files[1], 'r') as f:
    # 读取每个文件中的数据
    lines = f.readlines()
    # 将每行数据转换为浮点型，并添加到列表中
    data.append([float(x.strip())-0.41199 for x in lines])

with open(data_files[2], 'r') as f:
    # 读取每个文件中的数据
    lines = f.readlines()
    # 将每行数据转换为浮点型，并添加到列表中
    data.append([float(x.strip())+0.81995 for x in lines])

# 绘图
plt.figure(figsize=(10, 5))
colors = ['red', 'green', 'blue']

plt.plot(range(1, 301), data[0], color=colors[0], label='Faster-RCNN')
plt.plot(range(1, 301), data[1], color=colors[1], label='SSD')
plt.plot(range(1, 301), data[2], color=colors[2], label='YOLOv3')

plt.legend(loc='center right',prop={'size': 15})
plt.xlabel('epoch')
plt.ylabel('val')
plt.title('val_loss of 3 baseline models')
plt.grid(True)

plt.savefig("baseline1.png")
plt.show()
```
